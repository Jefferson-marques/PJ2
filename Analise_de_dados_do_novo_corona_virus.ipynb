# Visualiza√ß√£o atrav√©s de gr√°ficos

```
# n√∫mero de casos de acordo com o tempo
fig = px.bar(df_novo[df_novo['Country']=='China'], 
             x="Date", y="Count", color='Count', 
             title='N√∫mero de casos confirmados na China')
fig.update_traces(text=df_novo['Count'],textposition='outside')
fig.show()
```

```
# n√∫mero de casos de acordo com o a data
temp = df_novo[df_novo['Country']!='China'].groupby('Date').sum().reset_index()
fig = px.bar(temp,
             x="Date", y='Count', color='Count',
             title='N√∫mero de casos confirmados fora da China')
fig.update_traces(text=temp['Count'], textposition='outside')
fig.show()
```

```
#Outra maneira de ver os dados
fig = px.bar(df_novo[df_novo['Country']!='China'],
             y="Date", x="Count", color='Country', orientation='h', height=800,
             title='N√∫mero de casos confirmados fora da China')
fig.show()
```
